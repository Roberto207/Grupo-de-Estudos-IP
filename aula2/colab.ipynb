```python
from os import error
# ========================
# 1 - Tipos de Dados
# ========================

nome = "Ana"
print("String:", nome)

idade = 25
print("Inteiro:", idade)

altura = 1.68
print("Float:", altura)

estudando = True
print("Boolean:", estudando)


# ========================
# 1.1 Comandos interessantes para strings
# ========================

frase = "  Olá, bem-vindo ao grupo de estudos!  "

print("Tamanho da string:", len(frase))
print("Primeiros 10 caracteres:", frase[:10])
print("Caracteres pulando de 2 em 2:", frase[0:10:2])
print("Contar 'o':", frase.count('o'))
print("Posição de 'bem':", frase.find('bem'))
print("Tudo maiúsculo:", frase.upper())
print("Tudo minúsculo:", frase.lower())
print("Primeira letra maiúscula:", frase.capitalize())
print("Início de cada palavra maiúscula:", frase.title())
print("Remover espaços:", frase.strip())
print("Separar palavras:", frase.strip().split())
print("Juntar palavras com hífen:", "-".join(frase.strip().split()))
print("Encontrar 'bem' ignorando maiúsculas/minúsculas:", frase.lower().find('bem'))
```

```python
# ========================
# 2 - Estruturas de Condição
# ========================

x = 10
y = 5

if x > y:
    print("x é maior x")
elif x == y:
    print("x é igual a y")
else:
    print("x não é igual a nada")

if not False:
    print("Não False")

if x > 0 and y > 0:
    print("x e y são positivos")

if x > 0 or y < 0:
    print("x é positivo ou y é negativo")
```

```python
# ========================
# 3 - Laços de Repetição
# ========================

# For com range
for i in range(5):
    if i == 3:
        break  # sai do loop
    print("For range com break:", i)

# For com continue
for i in range(5):
    if i == 2:
        continue  # pula essa iteração
    print("For range com continue:", i)

# For percorrendo lista
frutas = ["maçã", "banana", "laranja"]
for fruta in frutas:
    print("Fruta:", fruta)

# While
contador = 0
while contador < 5:
    if contador == 3:
        break
    print("While contador:", contador)
    contador += 1
```

```python
#==================================
# 5 - Estruturas de Dados Compostos
# =================================

# ----- Listas -----
frutas = ["maçã", "banana", "laranja", "banana"]
frutas.append("pera")
frutas.remove("banana")
frutas.sort()
print("Lista de frutas:", frutas)
print("Tamanho da lista:", len(frutas))
print("Contar 'banana':", frutas.count("banana"))
print("Índice de 'laranja':", frutas.index("laranja"))
frutas.reverse()
print("Lista invertida:", frutas)
frutas2 = frutas.copy()
print("Cópia da lista:", frutas2)

# ----- Tuplas -----
coordenadas = (10, 20, 10)
print("Tupla:", coordenadas)
print("Tamanho da tupla:", len(coordenadas))
print("Contar 10:", coordenadas.count(10))
print("Índice do primeiro 20:", coordenadas.index(20))

# ----- Dicionários -----
usuario = {"nome": "Ana", "idade": 25, "email": "ana@email.com"}
print("Nome do usuário:", usuario["nome"])
usuario["idade"] = 26
print("Idade atualizada:", usuario["idade"])
print("Todas as chaves:", usuario.keys())
print("Todos os valores:", usuario.values())
print("Itens (chave-valor):", usuario.items())
print("Telefone do usuário (get):", usuario.get("telefone", "Não definido"))
usuario.update({"telefone": "11999999999"})
print("Usuário atualizado:", usuario)
```

```python
# ========================
# 6 - Tratamento de Erros (try e except)
# ========================

# Exemplo 1 - divisão
a = 10
b = int(input("digite 0 para erro,outros numeros para teste"))
try:
    resultado = a / b
except ZeroDivisionError:
    print("Erro: divisão por zero")
else:
    print("Resultado:", resultado)


# Exemplo 2 - abrir arquivo
try:
    with open("arquivo_que_nao_existe.txt", "r") as f:
        conteudo = f.read()
except FileNotFoundError:
    print("Erro: arquivo não encontrado")



# Exemplo genérico com finally
try:
    numero = int(input("digite uma letra"))  # vai gerar ValueError
except  ValueError as e:
    print("Erro genérico: algo deu errado")
else:
    print("Número:", numero)
finally:
    print("Bloco finally sempre executa") #finally sempre roda,independente dos erros e elses
```
