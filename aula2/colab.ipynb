```python
# ========================
# 1 - Tipos de Dados
# ========================

nome = "Ana"
print("String:", nome)

idade = 25
print("Inteiro:", idade)

altura = 1.68
print("Float:", altura)

estudando = True
print("Boolean:", estudando)


# ========================
# 1.1 Comandos interessantes para strings
# ========================

frase = "  Olá, bem-vindo ao grupo de estudos!  "

print("Tamanho da string:", len(frase))
print("Primeiros 10 caracteres:", frase[:10])
print("Caracteres pulando de 2 em 2:", frase[0:10:2])
print("Contar 'o':", frase.count('o'))
print("Posição de 'bem':", frase.find('bem'))
print("Tudo maiúsculo:", frase.upper())
print("Tudo minúsculo:", frase.lower())
print("Primeira letra maiúscula:", frase.capitalize())
print("Início de cada palavra maiúscula:", frase.title())
print("Remover espaços:", frase.strip())
print("Separar palavras:", frase.strip().split())
print("Juntar palavras com hífen:", "-".join(frase.strip().split()))
print("Encontrar 'bem' ignorando maiúsculas/minúsculas:", frase.lower().find('bem'))
```

```python
# ========================
# 2 - Estruturas de Condição
# ========================

x = 10
y = 5

if x > y:
    print("x é maior x")
elif x == y:
    print("x é igual a y")
else:
    print("x não é igual a nada")

if not False:
    print("Não False")

if x > 0 and y > 0:
    print("x e y são positivos")

if x > 0 or y < 0:
    print("x é positivo ou y é negativo")
```

```python
# ========================
# 3 - Laços de Repetição
# ========================

# For com range
for i in range(5):
    if i == 3:
        break  # sai do loop
    print("For range com break:", i)

# For com continue
for i in range(5):
    if i == 2:
        continue  # pula essa iteração
    print("For range com continue:", i)

# For percorrendo lista
frutas = ["maçã", "banana", "laranja"]
for fruta in frutas:
    print("Fruta:", fruta)

# While
contador = 0
while contador < 5:
    if contador == 3:
        break
    print("While contador:", contador)
    contador += 1
```

```python
#==================================
# 5 - Estruturas de Dados Compostos
# =================================


# ==================== LISTAS ====================
# Listas são coleções ORDENADAS, MUTÁVEIS e que permitem DUPLICATAS
# Ideal para: sequências de itens que precisam ser modificados
lista1 = list()
lista1.append([1,2,3,4,5,6])
print(lista1)
frutas = ["maçã", "banana", "laranja", "banana"]
frutas.append("pera")
frutas.remove("banana")
frutas.sort()
print("\nLista de frutas:", frutas)
print("\nTamanho da lista:", len(frutas))
print("\nContar 'banana':", frutas.count("banana"))
print("\nÍndice de 'laranja':", frutas.index("laranja"))
frutas.reverse()
print("\nLista invertida:", frutas)
frutas2 = frutas.copy()
print("\nCópia da lista:", frutas2)

# ==================== TUPLAS ====================
# Tuplas são coleções ORDENADAS, IMUTÁVEIS e que permitem DUPLICATAS
# Ideal para: dados que não devem mudar (coordenadas, configurações, chaves de dicionário)
# IMUTABILIDADE:
# coordenadas[0] = 30  # ❌ ERRO! TypeError: 'tuple' object does not support item assignment
tupla = tuple((3,2,1))
print(tupla)
coordenadas = (10, 20, 10)
print("\nTupla:", coordenadas)
print("\nTamanho da tupla:", len(coordenadas))
print("\nContar 10:", coordenadas.count(10))
print("\nÍndice do primeiro 20:", coordenadas.index(20))
# Mas objetos mutáveis dentro da tupla podem ser modificados:
tupla_com_lista = ([1, 2], 3)
tupla_com_lista[0].append(3)  # ✅ Funciona! Lista interna é mutável


# ==================== DICIONÁRIOS ====================
# Dicionários são coleções de pares CHAVE-VALOR, ORDENADAS, MUTÁVEIS
# Ideal para: mapeamentos, caches, contadores, estruturas de dados complexas

usuario = {"nome": "Ana", "idade": 25, "email": "ana@email.com"}

# ACESSO E MODIFICAÇÃO
print("\nNome do usuário:", usuario["nome"])       # O(1) médio. KeyError se chave não existe!
usuario["idade"] = 26                            # Modifica valor existente
usuario["cidade"] = "SP"                       # Adiciona nova chave automaticamente

# MÉTODOS DE CONSULTA SEGURA (evitam KeyError)
print("\nTelefone do usuário (get):", usuario.get("telefone", "Não definido"))
# .get(chave, default) retorna default se chave não existe (None se não especificado)

# ITERAÇÃO E VISUALIZAÇÃO
print("\nTodas as chaves:", usuario.keys())        # dict_keys(['nome', 'idade', 'email'])
print("\nTodos os valores:", usuario.values())     # dict_values(['Ana', 26, 'ana@email.com'])
print("\nItens (chave-valor):", usuario.items())   # dict_items([('nome', 'Ana'), ...])

# ATUALIZAÇÃO EM LOTE
usuario.update({"telefone": "11999999999"})      # Adiciona/atualiza múltiplos pares
# Também funciona: usuario.update(telefone="11999999999", cidade="SP")
print("\nUsuário atualizado:", usuario)

# OUTROS MÉTODOS ÚTEIS:
# usuario.setdefault("pais", "Brasil")  # Só adiciona se chave não existe
# usuario.pop("email")                   # Remove e retorna valor
# usuario.popitem()                      # Remove e retorna último par (LIFO desde Py 3.7)
# del usuario["idade"]                   # Remove chave
# "nome" in usuario                      # Verifica existência de chave (O(1))

tuplas = [(1, "um"), (2, "dois"), (3, "três")]

dicionario = dict(tuplas) 
print(dicionario)
```

```python
# ========================
# 6 - Tratamento de Erros (try e except)
# ========================

# Exemplo 1 - divisão
a = 10
b = int(input("digite 0 para erro,outros numeros para teste"))
try:
    resultado = a / b
except ZeroDivisionError:
    print("Erro: divisão por zero")
else:
    print("Resultado:", resultado)


# Exemplo 2 - abrir arquivo
try:
    with open("arquivo_que_nao_existe.txt", "r") as f:
        conteudo = f.read()
except FileNotFoundError:
    print("Erro: arquivo não encontrado")



# Exemplo genérico com finally
try:
    numero = int(input("digite uma letra"))  # vai gerar ValueError
except  ValueError as e:
    print("Erro genérico: algo deu errado")
else:
    print("Número:", numero)
finally:
    print("Bloco finally sempre executa") #finally sempre roda,independente dos erros e elses
```
